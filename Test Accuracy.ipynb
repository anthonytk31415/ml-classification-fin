```python
import matplotlib.pyplot as plt
import pandas as pd

# Create dataframe from the model performance data
data = {
    'Model': ['XGBoost', 'Random Forest', 'Neural Network', 'Logistic Regression'],
    'Score': [0.9077, 0.8965, 0.8925, 0.8905]
}

df = pd.DataFrame(data)

# Sort by score in descending order
df = df.sort_values('Score', ascending=False)

# Set up the plot
plt.figure(figsize=(6, 2))

# Create horizontal bar chart with different colors
colors = ['#2C7BB6', '#7FCDBB', '#FFC107', '#D7301F']
bars = plt.barh(df['Model'], df['Score'], color=colors)

# Set the x-axis limits to focus on the range 0.85 - 0.95
plt.xlim(0.85, 0.95)

# Add labels and title
plt.xlabel('Score')
plt.title('Model Performance Comparison - Test Accuracy')

# Add the actual values at the end of each bar
for i, bar in enumerate(bars):
    plt.text(bar.get_width() + 0.0005, bar.get_y() + bar.get_height()/2, 
             f'{df["Score"].iloc[i]:.4f}', 
             va='center')

# Add grid lines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
